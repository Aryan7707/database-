```python
import pyodbc
import json

def fetch_data_from_sql_server(server, database, username, password, table_name):
    # Create a connection string
    conn_str = (
        f"DRIVER={{SQL Server}};"
        f"SERVER={server};"
        f"DATABASE={database};"
        f"UID={username};"
        f"PWD={password}"
    )
    
    # Connect to the SQL Server
    try:
        conn = pyodbc.connect(conn_str)
        cursor = conn.cursor()
        
        # Query to fetch data from the specified table
        query = f"SELECT * FROM {table_name}"
        cursor.execute(query)
        
        # Fetch all rows and column names
        rows = cursor.fetchall()
        columns = [column[0] for column in cursor.description]
        
        # Convert rows to a list of dictionaries
        result = [dict(zip(columns, row)) for row in rows]
        
        # Convert the result to JSON format
        json_result = json.dumps(result, default=str)
        
        return json_result
    
    except pyodbc.Error as e:
        return json.dumps({"error": str(e)})
    
    finally:
        # Close the connection
        cursor.close()
        conn.close()

# Example usage
server = 'your_server'
database = 'your_database'
username = 'your_username'
password = 'your_password'
table_name = 'WORKER'

json_data = fetch_data_from_sql_server(server, database, username, password, table_name)
print(json_data)
```

```python
import mysql.connector
import json

def fetch_data_from_mysql(server, user, password, database, table_name):
    con = None
    cursor = None
    try:
        # Establish connection
        con = mysql.connector.connect(
            host=server,
            user=user,
            password=password,
            database=database,
            port=3306  # Adding the port to the connection parameters
        )
        cursor = con.cursor()
        
        # Define and execute the query
        query = f"SELECT * FROM {table_name}"
        cursor.execute(query)
        
        # Fetch all data
        columns = [column[0] for column in cursor.description]
        data = cursor.fetchall()
        
        # Convert to JSON format
        json_data = json.dumps([dict(zip(columns, row)) for row in data], default=str)
        
        return json_data
    except mysql.connector.Error as e:
        return json.dumps({"error": str(e)})
    finally:
        # Close the cursor and connection if they were opened
        if cursor:
            cursor.close()
        if con:
            con.close()

# Define connection parameters
server = '127.0.0.1'  # MySQL server address
user = 'root'
password = 'aryan7669'
database = 'ORG'
table_name = 'TITLE'

# Fetch data and print JSON
json_data = fetch_data_from_mysql(server, user, password, database, table_name)
print(json_data)
```
